```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {"User-Agent": "Mozilla/5.0"}
base_url = "https://www.sikayetvar.com"

data = []

# 5000 şikayet için yaklaşık 500 sayfa (her sayfada ~10 şikayet)
for page in range(1, 300):
    print(f"Sayfa {page} çekiliyor...")
    list_url = f"{base_url}/pegasus?page={page}"
    
    try:
        r = requests.get(list_url, headers=headers)
        r.encoding = "utf-8"
        soup = BeautifulSoup(r.text, "html.parser")
        
        links = [base_url + a["href"] for a in soup.select("h2 a")]
        
        for link in links:
            try:
                rr = requests.get(link, headers=headers)
                rr.encoding = "utf-8"
                ss = BeautifulSoup(rr.text, "html.parser")
                
                baslik = ss.find("h1", class_="complaint-detail-title").get_text(strip=True)
                
                icerik_div = ss.find("div", class_="complaint-detail-description")
                if icerik_div:
                    icerik_paragraflar = [p.get_text(strip=True) for p in icerik_div.find_all("p")]
                    icerik = " ".join(icerik_paragraflar)
                else:
                    icerik = ""
                
                data.append({"Başlık": baslik, "İçerik": icerik})
                print("Kaydedildi:", baslik)
                
                time.sleep(1)  # siteyi yormamak için bekle
            except Exception as e:
                print("Detay sayfasında hata:", e)
        
    except Exception as e:
        print("Liste sayfasında hata:", e)
    
    # Her 10 sayfada bir ara kayıt yap
    if page % 10 == 0:
        df = pd.DataFrame(data)
        df.to_excel("pegasus300_sikayetler_temp.xlsx", index=False)
        print(f"{page} sayfa sonrası geçici dosya kaydedildi.")
        time.sleep(5)  # küçük mola

# Tüm veriyi final Excel dosyasına kaydet
df = pd.DataFrame(data)
df.to_excel("pegasus300_complaints.xlsx", index=False)
print("Tüm şikayetler kaydedildi: pegasus300_complaints.xlsx")
```
